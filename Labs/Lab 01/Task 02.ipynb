## A) Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import io
```

## B) Load dataset from file

```python
data = pd.read_csv("cancer patient data sets.csv")
data.head(5)
```

## C) Find unique categories of the target variable

```python
possible_targets = ["Level", "Result", "Lung Cancer", "LUNG_CANCER", "lung_cancer"]
target_col = None
for c in possible_targets:
    if c in data.columns:
        target_col = c
        break
if target_col is None:
    target_col = data.columns[-1]

print("Target column:", target_col)
print("Unique categories:", data[target_col].unique())
```

## D) Display 50 from top, 30 from bottom, info, size, dimensions, memory

```python
display(data.head(50))
display(data.tail(30))

print("\nDataset info:")
buf = io.StringIO()
data.info(buf=buf)
print(buf.getvalue())

print("\nShape (rows, cols):", data.shape)
print("Total elements:", data.size)
print("Memory usage (bytes):", data.memory_usage(deep=True).sum())
```

## E) Number of columns and samples

```python
n_rows, n_cols = data.shape
print("Rows (samples):", n_rows)
print("Columns:", n_cols)
```

## F) Records per target category (and bar chart)

```python
vc = data[target_col].value_counts(dropna=False)
print(vc)

vc.plot(kind="bar")
plt.title(f"Counts per '{target_col}'")
plt.xlabel(target_col)
plt.ylabel("count")
plt.show()
```

## G) Extract rows 550–900 & features Gender → Coughing of Blood (loc)

```python
def col_slice(df, start, end):
    cols = list(df.columns)
    i0, i1 = cols.index(start), cols.index(end)
    return df.loc[:, cols[min(i0,i1):max(i0,i1)+1]]

subset_loc = data.loc[550:900]
g_df = subset_loc[col_slice(data, "Gender", "Coughing of Blood").columns]

g_df.to_csv("subset_loc.csv", index=False)
print("Saved subset_loc.csv")

desc = pd.DataFrame({
    "mean": g_df.mean(numeric_only=True),
    "median": g_df.median(numeric_only=True),
    "mode": g_df.mode(numeric_only=True).iloc[0],
    "std": g_df.std(numeric_only=True),
    "min": g_df.min(numeric_only=True),
    "max": g_df.max(numeric_only=True),
})
desc
```

## H) Last 20 rows, 5 features from Shortness of Breath → Frequent Cold (iloc)

```python
cols = list(data.columns)
i_start, i_end = cols.index("Shortness of Breath"), cols.index("Frequent Cold")
sel_cols = cols[min(i_start,i_end):max(i_start,i_end)+1][:5]

last20 = data.iloc[-20:, :]
h_df = last20[sel_cols]

h_df.to_csv("subset_iloc.csv", index=False)
print("Saved subset_iloc.csv")

h_desc = pd.DataFrame({
    "mean": h_df.mean(numeric_only=True),
    "median": h_df.median(numeric_only=True),
    "mode": h_df.mode(numeric_only=True).iloc[0],
    "std": h_df.std(numeric_only=True),
    "min": h_df.min(numeric_only=True),
    "max": h_df.max(numeric_only=True),
})
h_desc
```

## I) Selected indices & average Alcohol use

```python
idx = [110,220,360,440,656,778,202,889]
sample_i = data.loc[idx]

alcohol_col = [c for c in data.columns if "alcohol" in c.lower()][0]

avg_alcohol = sample_i[alcohol_col].astype(float).mean()
print("Alcohol column:", alcohol_col)
print("Average Alcohol use:", avg_alcohol)
```
