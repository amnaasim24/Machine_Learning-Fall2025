```python
import numpy as np
import pandas as pd
import io

np.random.seed(42)
n = 12
df1 = pd.DataFrame({
    "test_score": np.random.randint(20, 80, n),
    "writing_skills": np.random.randint(1, 10, n),
    "reading_skill": np.random.randint(1, 10, n),
    "study_hours": np.random.uniform(0.5, 4.0, n).round(2),
    "attendance": np.random.randint(50, 100, n),
})
df1["failed"] = ((df1["test_score"] < 40) | (df1["attendance"] < 60)).astype(int)
print("\nMethod 1: Dict of NumPy arrays")
print(df1)

np.random.seed(123)
s_test  = pd.Series(np.random.randint(20, 80, n), name="test_score")
s_write = pd.Series(np.random.randint(1, 10, n), name="writing_skills")
s_read  = pd.Series(np.random.randint(1, 10, n), name="reading_skill")
s_hrs   = pd.Series(np.random.uniform(0.5, 4.0, n).round(2), name="study_hours")
s_att   = pd.Series(np.random.randint(50, 100, n), name="attendance")

df2 = pd.DataFrame({
    "test_score": s_test,
    "writing_skills": s_write,
    "reading_skill": s_read,
    "study_hours": s_hrs,
    "attendance": s_att
})
df2["failed"] = ((df2["test_score"] < 40) | (df2["attendance"] < 60)).astype(int)
print("\nMethod 2: Dict of Series")
print(df2)

csv_text = """test_score,writing_skills,reading_skill,study_hours,attendance
35,3,4,1.2,55
78,7,6,3.5,92
62,5,6,2.1,88
41,4,5,1.8,60
28,3,2,0.9,58
44,5,5,2.2,72
53,6,6,2.5,80
31,2,3,1.0,57
69,6,7,2.7,85
59,5,5,2.4,74
"""
df3 = pd.read_csv(io.StringIO(csv_text))
df3["failed"] = ((df3["test_score"] < 40) | (df3["attendance"] < 60)).astype(int)
print("\nMethod 3: CSV String")
print(df3)
```
