```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv("cancer patient data sets.csv")

target_col = next((c for c in ["Level","Result","Lung Cancer","LUNG_CANCER","lung_cancer"] if c in data.columns), data.columns[-1])

if "Age" in data.columns:
    data["Age"].dropna().astype(float).hist(bins=20)
    plt.title("Distribution of Age")
    plt.show()

data[target_col].value_counts().plot(kind="bar")
plt.title("Target Distribution")
plt.show()

if "Obesity" in data.columns:
    data["Obesity"].dropna().astype(float).plot(kind="box")
    plt.title("Boxplot  Obesity")
    plt.show()

num = data.select_dtypes(include=[np.number])
if num.shape[1] >= 2:
    plt.imshow(num.corr(), cmap="coolwarm")
    plt.title("Correlation Matrix")
    plt.colorbar()
    plt.show()

if "Age" in data.columns:
    data["Age"].dropna().astype(float).hist(bins=20)
    plt.title("Distribution of Age")
    plt.show()

if "Frequent Cold" in data.columns:
    data["Frequent Cold"].dropna().astype(float).hist(bins=20)
    plt.title("Distribution of Frequent Cold")
    plt.show()

if "Obesity" in data.columns:
    data["Obesity"].dropna().astype(float).plot(kind="box")
    plt.title("Outliers in Obesity")
    plt.show()
```
